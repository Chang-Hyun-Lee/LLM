```python

import os
import openai
from pykrx import stock
from pykrx import bond

openai.api_key = os.getenv("OPENAI_API_KEY")

df = stock.get_market_ohlcv_by_date("20250701", "20250721", "005930")
df = df.reset_index()

csv_data = df[['날짜', '시가', '고가', '저가', '종가']].to_string(index=False)

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": "다음은 삼성전자의 날짜별 주가 데이터야. 이 데이터를 분석해서 보고해줘."
    },
    {
      "role": "user",
      "content": csv_data
    }
  ],
  temperature=0,
  max_tokens=1024,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

result = response.choices[0].message.content

print(result)
```
