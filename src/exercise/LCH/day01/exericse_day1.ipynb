```python
import os
import openai
import requests

openai.api_key = os.getenv("OPENAI_API_KEY")

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": """사용자가 입력한 질문에 맞는 캠핑장을 검색해서 추천하는 ChatGPT 어플리케이션을 작성하시오"""
    },
    {
      "role": "user",
      "content": "계곡이 있는 캠핑장을 두어개 추천해줘"
    }
  ],
  temperature=0,
  max_tokens=1048,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

def get_camping_sites():
    # API URL 및 인증키 설정
    url = "https://apis.data.go.kr/B551011/GoCamping/searchList"
    params = {
        'serviceKey': 'qVIFfwvBFGmr6r4QBzTX1wKVxP6KHx7UpB6SE2Iyho8%2Bn071XEofG4zjwTk39hDktTo%2Fii91yaPRg10Psldjlg%3D%3D',
        'numOfRows': 3,  # 추천할 캠핑장 수
        'pageNo': 1,
        'MobileOS': 'AND',
        'MobileApp': 'AppTest',
        'keyword' : '계곡'
    }

    # API 요청
    response = requests.get(url, params=params)

    # 응답 확인
    if response.status_code == 200:
        data = response.json()
        items = data.get('response', {}).get('body', {}).get('items', {}).get('item', [])
        
        # 캠핑장 정보 요약
        camping_sites = []
        for item in items:
            site_info = {
                'name': item.get('facltNm'),  # 캠핑장 이름
                'address': item.get('addr1'),  # 주소
                'phone': item.get('tel'),  # 전화번호
                'homepage': item.get('homepage'),  # 홈페이지
            }
            camping_sites.append(site_info)

        return camping_sites
    else:
        print("API 요청 실패:", response.status_code)
        return []

# 캠핑장 정보 출력
# if __name__ == "__main__":
#     camping_sites = get_camping_sites()
#     for idx, site in enumerate(camping_sites, start=1):
#         print(f"캠핑장 {idx}:")
#         print(f"이름: {site['name']}")
#         print(f"주소: {site['address']}")
#         print(f"전화번호: {site['phone']}")
#         print(f"홈페이지: {site['homepage']}")
#         print("-" * 30)

print(response.choices[0].message.content)
```
