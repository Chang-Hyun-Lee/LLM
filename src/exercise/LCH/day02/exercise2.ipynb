```python
import pandas as pd
from datetime import datetime
import os
import openai
from io import StringIO
from pykrx import stock
from pykrx import bond


def analyze_stock_data(stock_data, stock_name, start_date, end_date):
    # 데이터프레임으로 변환
    df = pd.DataFrame(stock_data)
    
    # 날짜 형식 변환
    df['날짜'] = pd.to_datetime(df['날짜'])
    
    # 시작일과 종료일에 따라 데이터 필터링
    mask = (df['날짜'] >= start_date) & (df['날짜'] <= end_date)
    filtered_data = df.loc[mask]
    
    # 데이터 분석
    if filtered_data.empty:
        return f"{stock_name}의 {start_date}부터 {end_date}까지의 데이터가 없습니다."
    
    # 기본 통계량
    analysis_report = {
        '종목': stock_name,
        '기간': f"{start_date} ~ {end_date}",
        '최고가': filtered_data['고가'].max(),
        '최저가': filtered_data['저가'].min(),
        '시가': filtered_data['시가'].iloc[0],
        '종가': filtered_data['종가'].iloc[-1],
        '거래량 합계': filtered_data['거래량'].sum(),
        '평균 등락률': filtered_data['등락률'].mean()
    }
    
    return analysis_report


openai.api_key = os.getenv("OPENAI_API_KEY")

df = stock.get_market_ohlcv("20220720", "20220810", "005930")
csv_buffer = StringIO()
df.to_csv(csv_buffer)

csv_data = csv_buffer.getvalue()
# print(csv_data)

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": f"{analyze_stock_data} 함수를 활용해줘"
    },
    {
      "role": "user",
      "content": "삼성전자 주가 데이터를 분석하는 어플리케이션을 function(tool)으로 수정하여, 다른 종목의 데이터를 정해진 내용대로(종목, 시작날짜, 종료날짜) 가져와서 답하는 어플리케이션을 작성하시오"
    }
  ],
  temperature=0,
  max_tokens=1024,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```
