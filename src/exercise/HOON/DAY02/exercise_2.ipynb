```python
import os
ServiceKey = os.getenv("FOOD_API_KEY")
```

```python
from urllib.parse import quote

KeyWord = quote("지역")
print(KeyWord)
```

```python
url = f"http://apis.data.go.kr/6260000/BusanTblFnrstrnStusService/getTblFnrstrnStusInfo?serviceKey={ServiceKey}&numOfRows=100&pageNo=1&resultType=json"
print(url)
```

```python
import requests

def get_url_content(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            return response.text
        else:
            response.raise_for_status()

    except requests.exceptions.RequestException as e:
        print(f"Request Exception: {e}")
        return None

result = get_url_content(url)
if result is not None:
    print(result)
else:
    print("Failed to retrieve URL content.")
```

```python
import json

data = json.loads(result)
data
```

```python
sites = data['response']['body']['items']['item']
```

```python
for site in sites:
    print(site['bsnsNm'] + ":" + site['menu'])
```

```python
question = "한우 파는 곳 추천해줘."
```

```python
import os
import openai

openai.api_key = os.getenv("OPENAI_API_KEY")

response = openai.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {
      "role": "system",
      "content": "다음 질문에서 가장 중요한 키워드 단어 하나만 뽑아줘."
    },
    {
      "role": "user",
      "content": question
    }
  ],
  temperature=0,
  max_tokens=1024,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```

```python
KeyWord = response.choices[0].message.content
```

```python
url = f"http://apis.data.go.kr/6260000/BusanTblFnrstrnStusService/getTblFnrstrnStusInfo?serviceKey={ServiceKey}&bsnsNm={KeyWord}&numOfRows=100&pageNo=1&resultType=json"
print(url)
```

```python
result = get_url_content(url)

import json

data = json.loads(result)
sites = data['response']['body']['items']['item']

context = ''
i = 1;
for site in sites:
    context = context + str(i) + ") " + site['bsnsNm'] + ":" + site['addrRoad'] + "\n"
    i = i + 1
context
```

```python
import os
import openai

openai.api_key = os.getenv("OPENAI_API_KEY")

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": f"주어진 문단을 보고 다음 질문에 답해줘: {question}"
    },
    {
      "role": "user",
      "content": context
    }
  ],
  temperature=0,
  max_tokens=1024,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```
