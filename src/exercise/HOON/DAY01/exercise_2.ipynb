실습: 사용자가 입력한 질문에 맞는 캠핑장을 검색해서 추천하는 ChatGPT 어플리케이션을 작성하시오.
예) 계곡이 있는 캠핑장을 두어개 추천해줘.

```python
# https://www.data.go.kr/data/15101933/openapi.do 고캠핑 API
```

```python
import os
ServiceKey = os.getenv("GOCAMP_API_KEY")
```

```python
from urllib.parse import quote

KeyWord = quote("계곡")
print(KeyWord)
```

```python
url = f"http://apis.data.go.kr/B551011/GoCamping/searchList?serviceKey={ServiceKey}&keyword={KeyWord}&numOfRows=10&pageNo=1&MobileOS=ETC&MobileApp=TestApp&_type=json"
print(url)
```

```python
import requests

def get_url_content(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            return response.text
        else:
            response.raise_for_status()

    except requests.exceptions.RequestException as e:
        print(f"Request Exception: {e}")
        return None

result = get_url_content(url)
if result is not None:
    print(result)
else:
    print("Failed to retrieve URL content.")
```

```python
import json

data = json.loads(result)
data
```

```python
sites = data['response']['body']['items']['item']
```

```python
for site in sites:
    print(site['facltNm'] + ":" + site['intro'])
```

```python
question = "수영장이 있는 캠핑장을 두어개 추천해줘."
```

```python
import os
import openai

openai.api_key = os.getenv("OPENAI_API_KEY")

response = openai.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {
      "role": "system",
      "content": "다음 질문에서 가장 중요한 키워드 단어 하나만 뽑아줘."
    },
    {
      "role": "user",
      "content": question
    }
  ],
  temperature=0,
  max_tokens=1024,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```

```python
KeyWord = response.choices[0].message.content
```

```python
url = f"http://apis.data.go.kr/B551011/GoCamping/searchList?serviceKey={ServiceKey}&keyword={KeyWord}&numOfRows=10&pageNo=1&MobileOS=ETC&MobileApp=TestApp&_type=json"
```

```python
result = get_url_content(url)

import json

data = json.loads(result)
sites = data['response']['body']['items']['item']

context = ''
i = 1;
for site in sites:
    context = context + str(i) + ") " + site['facltNm'] + ":" + site['intro'] + "\n"
    i = i + 1
context
```

```python
import os
import openai

openai.api_key = os.getenv("OPENAI_API_KEY")

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": f"주어진 문단을 보고 다음 질문에 답해줘: {question}"
    },
    {
      "role": "user",
      "content": context
    }
  ],
  temperature=0,
  max_tokens=1024,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```
