실습: 삼성전자의 주가 데이터를 가져와서 분석하는 ChatGPT 어플리케이션을 작성하시오(힌트: pykrx 패키지)

```python
from pykrx import stock
from pykrx import bond

print("---- 삼성전자 주가 데이터 가져오기")

name_code = "005930"
name = stock.get_market_ticker_name(name_code)
print(name)
```

```python
prices2025 = stock.get_market_ohlcv("20250101", "20250721", name_code)
print(prices2025)

pricesfrom2020 = stock.get_market_ohlcv("20200101", "20250721", name_code)
print(pricesfrom2020)

pricesfrom2023 = stock.get_market_ohlcv("20230101", "20250721", name_code)
print(pricesfrom2023)


pricesfor2024 = stock.get_market_ohlcv("20240101", "20241231", name_code)
print(pricesfor2024)
```

```python
df1 = pricesfor2024
print(df1)
```

```python
import os
import openai

openai.api_key = os.getenv("OPENAI_API_KEY")

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": """
넌 주식 데이터 분석 전문 보조 AI야. 사용자가 제공하는 주식 데이터를 분석하여 인사이트를 제공하고, 트렌드를 설명하며, 경향도 같이 설명해. 
차트 분석에 대한 기초 제안을 할 수 있어.
"""
    }, 
    {
      "role": "user",
      "content": f"""
내가 주는건 삼성전자 2024년의 주가 데이터야.
{df1}

이 데이터를 가지고 아래를 분석해줘.
1. 해당 기간 동안 삼성전자 주가 트렌드 분석
2. 가장 높은 종가와 낮은 종가와 날짜 를 json 형태로 알려줘.
3. 거래량이 많았던 15개의 거래일과 그때의 종가를 나타내줘.
"""
    },
  ],
  temperature=0,
  max_tokens=2024,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```
