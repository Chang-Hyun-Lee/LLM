```python
import os
import pandas as pd
from pykrx import stock
import matplotlib.pyplot as plt 
from datetime import datetime, timedelta
# 환경 변수에서 OpenAI API 키를 가져옵니다.
# 환경 변수에 API 키를 설정해야 합니다.
import openai

# OpenAI API 키 설정
# 예: os.environ["OPENAI_API_KEY"] = "your-api-key"

end_date = datetime.today()
start_date = end_date - timedelta(days=365)
df_samsung = stock.get_market_ohlcv_by_date(
    start_date.strftime("%Y%m%d"),
    end_date.strftime("%Y%m%d"),
    "005930"
)

# 데이터 간단 분석: 종가 시계열 그래프
df_samsung['종가'].plot(title='삼성전자 최근 1년간 종가 추이', figsize=(10, 5))
plt.xlabel('날짜')
plt.ylabel('종가')
plt.show()

# 데이터프레임의 기본 통계 정보 출력
print(df_samsung.describe())

openai.api_key = os.getenv("OPENAI_API_KEY")

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": "삼성전자 주식 경향을 설명해줘."
    },
    {
      "role": "user",
      "content": "그 머스마가 니 마음에 안등다 그 카드나? 계속 꼬시보지?"
    }
  ],
  temperature=0,
  max_tokens=256,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```
