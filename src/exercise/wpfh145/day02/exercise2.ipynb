```python
# 실습N #1: 엔비디아의 주가 데이터를 가져와서 분석하는 어플리케이션을 작성하시오.
import os
import openai
import yfinance as yf
from io import StringIO

openai.api_key = os.getenv("OPENAI_API_KEY")

ticker = yf.Ticker("NVDA") # nvidia 주가
df = ticker.history(period="5y")  # 전체 기간의 OHLCV 데이터

# CSV로 변환
csv_buffer = StringIO()
df.to_csv(csv_buffer)

ohlcv_csv_data = csv_buffer.getvalue()

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": f"{ohlcv_csv_data}는 Nvidia의 날짜별 주가 데이터야. 이 데이터를 분석해서 보고해줘."
    }
  ],
  temperature=0,
  max_tokens=1024,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```
