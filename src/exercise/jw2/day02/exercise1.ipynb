```python
import yfinance as yf

ticker = yf.Ticker("NVDA") # nvidia 주가
df = ticker.history(period="max")  # 전체 기간의 OHLCV 데이터

import os
import openai

openai.api_key = os.getenv("OPENAI_API_KEY")

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": "다음은 엔비디아의 전체기간 주가 데이터야. 이 데이터를 분석해서 보고해줘."
    },
    {
      "role": "user",
      "content": f"정렬된 {df}에서 open을 기준으로 상위 5개의 날짜와 주가를 추출해줘"
    }
  ],
  temperature=0,
  max_tokens=1024,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```

```python
# 카카오 API 활용
# 실습N #2: 사용자가 입력한 지역과 질문에 맞는 지역 음식점을 검색해서 추천하는 어플리케이션을 작성하시오.
# 예) 판교에 있는 괜찮은 이탈리안 식당을 몇개 추천해줄래?

from urllib.parse import quote
import requests
import json

openai.api_key = os.getenv("OPENAI_API_KEY")


API_KEY = "70200e2a88cc0ab7ee1632c35d8aac0b"

url = "https://dapi.kakao.com/v2/local/search/keyword.json"

headers = {
    "Authorization": f"KakaoAK {API_KEY}"
}

params = {
    "query": "식당"
}

response_api = requests.get(url, headers=headers, params=params)

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": f"{response_api}는 경기도의 모범 음식점 데이터야. 이 데이터를 분석해서 사용자가 원하는 음식점을 추천해줘."
    },
    {
        "role" : "user",
        "content" : "판교 괜찮은 식당 몇개를 추천해줄래?"
    }
  ],
  temperature=0,
  max_tokens=1024,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```
