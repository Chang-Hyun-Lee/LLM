```python
# user 없이 system prompt만으로 실행

import os
import openai

openai.api_key = os.getenv("OPENAI_API_KEY")

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": "종목별로 뭐든 분석해줘."
    }
  ],
  temperature=0,
  max_tokens=256,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```

```python
# user 없이 system prompt만으로 실행
# df : get_market_ohlcv 함수는 시작일/종료일/티커 세 개의 파라미터를 입력받아 OHLCV를 일자별로 정렬하여 DataFrame으로 반환합니다.

import os
from pykrx import stock
import openai

openai.api_key = os.getenv("OPENAI_API_KEY")

df = stock.get_market_ohlcv("20220720", "20220810", "005930")

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": f"최근 5년간의 {df}의 거래량을 분석해줘."
    }
  ],
  temperature=0,
  max_tokens=256,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```
