```python
# 주가 연습

import os
import openai
from pykrx import stock

openai.api_key = os.getenv("OPENAI_API_KEY")
df = stock.get_market_trading_value_by_date("20250701", "20250720", "005930")

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": "주어진 정보를 바탕으로 주가 분석을 해줘."
    },
    {
      "role": "user",
      "content": df.head().to_string()
    }
  ],
  temperature=0,
  max_tokens=1024,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```
