```python
# 엔비디아 (NVDA) 티커 심볼
nvidia_ticker = "NVDA"

# 주가 데이터 가져오기 (2022년 7월 20일 ~ 2022년 8월 10일)
# yfinance는 'YYYY-MM-DD' 형식의 날짜를 사용해요!
df_nvidia = yf.download(nvidia_ticker, start="2022-07-20", end="2022-08-11") # end 날짜는 가져오고 싶은 마지막 날짜 다음날로 지정해야 해요!
```

```python
df_nvidia
```

```python
from io import StringIO
csv_buffer = StringIO()
df_nvidia.to_csv(csv_buffer)

csv_data = csv_buffer.getvalue()
print(csv_data)
```

```python
import os
import openai

openai.api_key = os.getenv("OPENAI_API_KEY")

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": "다음은 엔비디아의 날짜별 주가 데이터야. 이 데이터를 분석해서 보고해줘."
    },
    {
      "role": "user",
      "content": csv_data
    }
  ],
  temperature=0,
  max_tokens=1024,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```
