```python
import gradio as gr
import openai
import os

# âœ… OpenAI API í‚¤ ì„¤ì •
openai.api_key = os.getenv("OPENAI_API_KEY")

# âœ… ìŠ¤íŠ¸ë¦¬ë° ì±— ì‘ë‹µ í•¨ìˆ˜
def chat_with_openai(message, history):
    messages = [{"role": "system", "content": "AssistantëŠ” ì¹œì ˆí•˜ê³  ìœ ìµí•˜ê²Œ ëŒ€í™”í•©ë‹ˆë‹¤."}]
    for user_msg, bot_msg in history:
        messages.append({"role": "user", "content": user_msg})
        messages.append({"role": "assistant", "content": bot_msg})
    messages.append({"role": "user", "content": message})

    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        stream=True,
        temperature=0.5,
        max_tokens=1024
    )

    partial_text = ""
    for chunk in response:
        if chunk.choices[0].delta and chunk.choices[0].delta.content:
            partial_text += chunk.choices[0].delta.content
            yield history + [(message, partial_text)], ""  # âœ… ì…ë ¥ì°½ ë¹„ìš°ê¸° í¬í•¨

# âœ… Gradio ì¸í„°í˜ì´ìŠ¤ êµ¬ì„±
with gr.Blocks() as demo:
    chatbot = gr.Chatbot()
    msg = gr.Textbox(label="ğŸ’¬ ë©”ì‹œì§€ë¥¼ ì…ë ¥í•˜ì„¸ìš”")
    clear = gr.Button("ğŸ§¹ ì´ˆê¸°í™”")

    def reset():
        return [], ""

    msg.submit(chat_with_openai, [msg, chatbot], [chatbot, msg], show_progress=True)  # âœ… ì¶œë ¥ì— msgë„ í¬í•¨
    clear.click(reset, outputs=[chatbot, msg])

# âœ… ì‹¤í–‰
demo.launch()
```
